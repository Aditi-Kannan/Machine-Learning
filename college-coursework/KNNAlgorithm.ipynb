```python
#importing libraries
from math import sqrt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#dataset
dataset = sns.load_dataset("iris")
#dataset.species.unique() => ['setosa', 'versicolor', 'virginica']
dataset['species']= dataset["species"].map({"setosa": 1, "versicolor": 2,"virginica":3})
display(dataset)
print("\n\n\n\n")
training_data, testing_data = train_test_split(dataset, test_size=0.2, random_state=25)
testing_data1  = testing_data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

results = pd.DataFrame()
results['Observed'] = testing_data['species']
results['Predicted']= np.NaN 
results = results.reset_index()

display(training_data)
print("\n\n\n\n")
display(testing_data1)

training_data1  = training_data.copy(deep=True).values.tolist()
testing_data2  = testing_data1.copy(deep=True).values.tolist()
```

```python
# calculate the Euclidean distance between two vectors
def euclidean_distance(row1, row2):
	distance = 0.0
	for i in range(len(row1)-1):
		distance += (row1[i] - row2[i])**2
	return sqrt(distance)

# Locate the most similar neighbors
def get_neighbors(train, test_row, num_neighbors):
	distances = list()
	for train_row in train:
		dist = euclidean_distance(test_row, train_row)
		distances.append((train_row, dist))
	distances.sort(key=lambda tup: tup[1])
	neighbors = list()
	for i in range(num_neighbors):
		neighbors.append(distances[i][0])
	return neighbors

# Make a classification prediction with neighbors
def predict_classification(train, test_row, num_neighbors):
	neighbors = get_neighbors(train, test_row, num_neighbors)
	output_values = [row[-1] for row in neighbors]
	prediction = max(set(output_values), key=output_values.count)
	return prediction
```

```python
k = int(input())
```

```python
for ii in range(0,len(testing_data2)):
    results['Predicted'][ii] = predict_classification(training_data1, testing_data2[ii], k)


results1  = results.copy(deep=True)

results['Observed']= results["Observed"].map({1:"setosa", 2:"versicolor",3:"virginica"})
results['Predicted']= results["Predicted"].map({1:"setosa", 2:"versicolor",3:"virginica"})

results
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
print("Confusion matrix:\n",confusion_matrix(results1['Observed'], results1['Predicted']))


# Accuracy
from sklearn.metrics import accuracy_score
print("\n\naccuracy score :\n",accuracy_score(results1['Observed'], results1['Predicted']))


from sklearn.metrics import f1_score
print("\n\nf1 score:\n",f1_score(results1['Observed'], results1['Predicted'], average=None),"\n\n\n")


from sklearn.metrics import classification_report
print(classification_report(results1['Observed'], results1['Predicted'])) #, target_names=target_names))
```
