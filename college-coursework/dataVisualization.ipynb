```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
import statistics
import pandas as pd
import joypy as jp
import random
```

```python
data = sns.load_dataset("tips")
print(data)
```

```python
# Creating histogram
nbins=10
fig, axs = plt.subplots(1, 1,figsize =(10,8),tight_layout = True)
axs.hist('total_bill',bins=nbins,data=data,color="pink",ec="lightblue",lw=2)
plt.title('histogram')
plt.xlabel('Total bill ($)')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter('total_bill', 'tip',data=data,color="red",s=5)
plt.xlabel('Total bill ($)')
plt.ylabel('tip ($)')
plt.show()
```

```python
fig = plt.figure(figsize =(10, 8))

d=data.loc[:,["total_bill","tip"]]

plt.boxplot(d)
plt.show()
```

```python
fig = plt.figure(figsize =(10, 8))
plt.violinplot(d)
plt.show()
```

```python
fig = plt.figure(figsize =(10, 8))
d=data['sex'].value_counts()
mylabels = ["Male", "Female"]
plt.pie(d, labels = mylabels, startangle = 90)
plt.show()
```

```python
fig = plt.figure(figsize =(10, 8))
df = data.loc[:,["total_bill","tip"]]
sns.heatmap(df.loc[0:10], annot=True, annot_kws={"size": 15})
```

```python
fig = plt.figure(figsize =(10, 8))
df = data.loc[:,"total_bill"]
sns.distplot(df)
plt.show()
```

```python

sns.jointplot(x = "total_bill", y = "tip",
              kind = "reg", data = data)

plt.show()
```

```python
sns.lmplot(x="total_bill",y="tip",data=data)
```

```python
fig = plt.figure(figsize =(10, 8))
sns.lineplot(x="total_bill", y="tip", data=data.loc[0:20])
plt.show()
```

```python
sns.pairplot(data, hue ='sex',palette='spring')
# to show
plt.show()
```

```python
fig = plt.figure(figsize =(10, 8))
sns.countplot(data=data, y="sex", hue="smoker")
```

```python
fig = plt.figure(figsize =(10, 8))
sns.barplot(pd.get_dummies(data['sex']))
```

```python
fig = plt.figure(figsize =(10, 8))
fig, ax = jp.joyplot(data.loc[0:10], by = "sex")

plt.show()
```
